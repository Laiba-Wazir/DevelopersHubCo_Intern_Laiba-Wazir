# *Visualization using Seaborn*

○ Scatter plot to analyze relationships between variables.

```python
new_df = df.drop(columns=['Id'])   # drop the ID column, as it does not affect the other columns in plotting.
```

```python
scatter = sns.scatterplot(data=new_df)    # creating a scatter plot using seaborn.
scatter.set_title('Scatter Plot Of Iris Dataset')   ## set title of Scatter plot.
scatter.set_xlabel('Frequency (f)')     # set label of x-axis.
scatter.set_ylabel('Centi-Meters (cm)')   # set label of y-axis.
scatter.grid(True)
```

○ Histogram to examine data distribution.

```python
hist = sns.histplot(data=new_df, element='step', kde=True)  # "element='step'" draws step-style bars, and "kde=True" adds a kernel density curve.
hist.set_title('Histogram Of Iris Dataset')   # set title of Histogram.
hist.set_xlabel('Centi-Meters (cm)')    # set label of x-axis.
hist.set_ylabel('Frequency (f)')    # set label of y-axis.
```

○ Box plot to detect outliers and spread of values.

```python
box = sns.boxplot(data=new_df)  # creating a Box plot using seaborn.
box.set_title('Box Plot Of Iris Dataset')   ## set title of Box plot.
box.set_ylabel('Centi-Meters (cm)')   # set label of y-axis.
box.grid(True)
```

# *Visualization using Matplotlib*

○ Scatter plot to analyze relationships between variables

```python
# set figure size for first scatter plot
plt.figure(figsize=(5, 4))

# Group data by species and plot each group separately
for species, group in df.groupby('Species'):
    plt.scatter(
        group['SepalLengthCm'],  # x-axis
        group['SepalWidthCm'],   # y-axis
        label=species     # label for each species
    )

# Add axis labels, title and legend
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width by Species')
plt.legend()


# set figure size for the second scatter plot
plt.figure(figsize=(5, 4))

# Group data by species and plot each group separately
for species, group in df.groupby('Species'):
    plt.scatter(
        group['PetalLengthCm'], # x-axis
        group['PetalWidthCm'],   # y-axis
        label=species     # label for each species
    )

# Add axis labels, titles and legend
plt.xlabel('Patel Length (cm)')
plt.ylabel('Patel Width (cm)')
plt.title('Patel Length vs Patel Width by Species')
plt.legend()
```

○ Histogram to examine data distribution.

```python
# create histogram for Petal length
plt.figure(figsize=(3, 3))    # set figure size
hist = plt.hist(df['PetalLengthCm'], bins=20, color='skyblue', edgecolor='black')
plt.title('For PetalLengthCm')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency (f)')


# create histogram for Petal width
plt.figure(figsize=(3, 3))    # set figure size
box = plt.hist(df['PetalWidthCm'], bins=20, color='hotpink', edgecolor='black')
plt.title('For PetalWidthCm')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency (f)')

# create histogram for Sepal length
plt.figure(figsize=(3, 3))    # set figure size
hist = plt.hist(df['SepalLengthCm'], bins=20, color='lightgreen', edgecolor='black')
plt.title('For SepalLengthCm')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency (f)')

# create histogram for Sepal width
plt.figure(figsize=(3, 3))    # set figure size
box = plt.hist(df['SepalWidthCm'], bins=20, color='orange', edgecolor='black')
plt.title('For SepalWidthCm')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency (f)')
```

○ Box plot to detect outliers and spread of values.

```python
box_var = [df['SepalLengthCm'], df['SepalWidthCm'], df['PetalLengthCm'], df['PetalWidthCm']]
box2 = plt.boxplot(box_var)
plt.xlabel('Features')
plt.title('Box Plot Of PetalLengthCm')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
```
